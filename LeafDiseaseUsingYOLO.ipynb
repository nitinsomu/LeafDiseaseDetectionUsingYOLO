```python
import os

def count_files_in_folder(folder_path):
    total_files = 0
    
    for item in os.listdir(folder_path):
        item_path = os.path.join(folder_path, item)
        if os.path.isfile(item_path):
            total_files += 1
        elif os.path.isdir(item_path):
            total_files += count_files_in_folder(item_path)
    
    return total_files
```

```python
folder_path = r"C:\Users\nitin\Desktop\Semester 7\Project 1\LeafDiseaseDetectionUsingYOLOv8\plant_disease_dataset\train"
print(f"Total train images: {count_files_in_folder(folder_path)}")
```
